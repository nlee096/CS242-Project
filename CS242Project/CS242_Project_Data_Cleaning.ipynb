Combine 700 with results_combined (Scraped data from IDs)

```python
df = pd.read_json('700_list.json')
rc = pd.read_json('results_combined.json')
```

```python
rc.head()
```

```python
columns=['plot', 'ratings', 'director', 'writers', 'stars', 'img', 'descriptors', 'url']
rc = rc.drop(columns, axis=1)
rc.head(5)
```

Use lambda function to convert title list to just title

```python
rc['title'] = rc['title'].apply(lambda x: x[0])
```

Do the same for synopsis (however it fails since the json read the brackets as a string)

```python
rc['synopsis'] = rc['synopsis'].apply(lambda x: x[0])
```

Instead, change synopsis to string format then replace brackets with ''

```python
rc['synopsis']=rc['synopsis'].astype(str)
rc['synopsis']=rc['synopsis'].str.replace(']','')
rc['synopsis']=rc['synopsis'].str.replace('[','')
rc.head()
```

Drop duplicates, then merge

```python
rc = rc.drop_duplicates(subset=['title'])
test_df = df.merge(rc, how='left', left_on='primaryTitle', right_on='title')
```

```python
test_df.head()
```

Create new synopsis column then add one of the two synopsis (from list json and ids json) into the column to be used as the default (list json overrides ids json here)

```python
test_df["synopsis"] = None
test_df.head()
```

```python
test_df['synopsis'] = test_df.apply(lambda x: x.synopsis_x if x.synopsis_x != None else x.synopsis_y, axis=1)
```

```python
test_df.head()
```
